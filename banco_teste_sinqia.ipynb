# [dbo].[InserirCep]

```sql
/****** Object:  StoredProcedure [dbo].[InserirCep]    Script Date: 12/03/2024 18:18:16 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[InserirCep]
    @Cep VARCHAR(9),
    @Endereco VARCHAR(100),
    @NumeroInicial INT,
    @NumeroFinal INT,
    @BairroDistrito VARCHAR(50),
    @Cidade VARCHAR(50),
    @UF VARCHAR(2),
    @Status INT OUTPUT
AS
BEGIN
    SET NOCOUNT ON;

    -- Verificar se o CEP já existe na tabela
    IF NOT EXISTS (SELECT 1 FROM CEP WHERE Cep = @Cep)
    BEGIN
        -- Se o CEP não existe, realizar a inserção
        INSERT INTO CEP(Cep, Endereco, NumeroInicial, NumeroFinal, BairroDistrito, Cidade, UF)
        VALUES (@Cep, @Endereco, @NumeroInicial, @NumeroFinal, @BairroDistrito, @Cidade, @UF);

        -- Definir o código de status como sucesso (por exemplo, 0)
        SET @Status = 0;
    END
    ELSE
    BEGIN
        -- Se o CEP já existe, definir o código de status como indicativo de CEP duplicado (por exemplo, 1)
        SET @Status = 1;
    END;
END;
GO
```
